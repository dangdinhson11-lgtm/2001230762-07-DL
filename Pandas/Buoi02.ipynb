## 3. Sinh viên tìm hiểu và viết chương trình sử dụng các hàm sau numpy.greater, numpy.greater_equal, numpy.less, và numpy.less_equal để so sánh 2 mảng 1 chiều, 2 chiều.

```python
# Mảng 1 chiều
arr1_1d = np.array([1, 2, 3, 4, 5])
arr2_1d = np.array([5, 3, 2, 4, 1])

print("Mảng 1D 1:", arr1_1d)
print("Mảng 1D 2:", arr2_1d)
print("arr1_1d > arr2_1d:", np.greater(arr1_1d, arr2_1d))
print("arr1_1d >= arr2_1d:", np.greater_equal(arr1_1d, arr2_1d))
print("arr1_1d < arr2_1d:", np.less(arr1_1d, arr2_1d))
print("arr1_1d <= arr2_1d:", np.less_equal(arr1_1d, arr2_1d))

print("\n" + "-"*30 + "\n")

# Mảng 2 chiều
arr1_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr2_2d = np.array([[3, 2, 1], [6, 5, 4]])

print("Mảng 2D 1:\n", arr1_2d)
print("Mảng 2D 2:\n", arr2_2d)
print("arr1_2d > arr2_2d:\n", np.greater(arr1_2d, arr2_2d))
print("arr1_2d >= arr2_2d:\n", np.greater_equal(arr1_2d, arr2_2d))
print("arr1_2d < arr2_2d:\n", np.less(arr1_2d, arr2_2d))
print("arr1_2d <= arr2_2d:\n", np.less_equal(arr1_2d, arr2_2d))
```

```python
np.random.seed(3) # for reproducibility
array_8 = np.random.randint(0, 21, size=20)
print("Mảng gốc:", array_8)

# Tạo một bản sao để không sửa đổi mảng gốc trực tiếp
modified_array_8 = array_8.copy()

# Xác định các phần tử trong khoảng [9, 15]
condition = (modified_array_8 >= 9) & (modified_array_8 <= 15)

# Đổi dấu các phần tử thỏa mãn điều kiện
modified_array_8[condition] = -modified_array_8[condition]

print("Mảng sau khi đổi dấu:", modified_array_8)
```

## 15. Viết chương trình Pandas để cộng, trừ, nhân, chia 2 Series : [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]

## 16. Viết chương trình Pandas chuyển đổi cột đầu tiên thành Series (Tìm hiểu thuộc tính ix của pandas)

```python
data = {'col1': [10, 20, 30, 40],
        'col2': [100, 200, 300, 400],
        'col3': [1, 2, 3, 4]}
df_16 = pd.DataFrame(data)
print("DataFrame gốc:\n", df_16)

# Thuộc tính .ix đã không còn được dùng. Thay vào đó, ta có thể chọn cột trực tiếp hoặc dùng .iloc
# Cách 1: Chọn cột bằng tên
series_from_col1 = df_16['col1']
print("\nCột đầu tiên ('col1') dưới dạng Series:\n", series_from_col1)
print("Kiểu dữ liệu của series_from_col1:", type(series_from_col1))

# Cách 2: Sử dụng .iloc để chọn cột theo vị trí (cột đầu tiên là index 0)
series_from_col_index0 = df_16.iloc[:, 0]
print("\nCột đầu tiên (index 0) dưới dạng Series (dùng .iloc):\n", series_from_col_index0)
print("Kiểu dữ liệu của series_from_col_index0:", type(series_from_col_index0))
```

## 19. Viết chương trình Pandas tìm những phần tử trong Series x mà không có trong Series y (Tìm hiểu hàm isin của pandas)

```python
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

print("sr1:\n", sr1)
print("sr2:\n", sr2)

# Tìm các phần tử trong sr1 không có trong sr2
items_not_in_sr2 = sr1[~sr1.isin(sr2)]
print("\nItems of sr1 not present in sr2:\n", items_not_in_sr2)
```

## 20. Tìm hiểu phép toán numpy.union1d, numpy.intersect1d (hội, giao) và viết chương trình minh họa 2 hàm này trên 2 Series x, y ngẫu nhiên có 20 phần tử

```python
np.random.seed(9) # for reproducibility
series_x_20 = pd.Series(np.random.randint(0, 30, size=20))
series_y_20 = pd.Series(np.random.randint(15, 45, size=20))

print("Series x (20 phần tử):\n", series_x_20)
print("\nSeries y (20 phần tử):\n", series_y_20)

# Sử dụng np.union1d cho phép hội
union_result = np.union1d(series_x_20, series_y_20)
print("\nPhép hội (Union) của Series x và y:\n", union_result)

# Sử dụng np.intersect1d cho phép giao
intersect_result = np.intersect1d(series_x_20, series_y_20)
print("\nPhép giao (Intersection) của Series x và y:\n", intersect_result)
```

## 21. Viết chương trình Pandas thêm series y vào series x theo chiều dọc (vertical), chiều ngang (horizontal) (Tìm hiểu hàm append và concat của pandas)

```python
series_x_21 = pd.Series(range(10))
series_y_21 = pd.Series([chr(ord('p') + i) for i in range(10)], index=range(10))

print("Series x:\n", series_x_21)
print("\nSeries y:\n", series_y_21)

# Thêm series y vào series x theo chiều dọc (vertical)
# append đã bị deprecated, sử dụng pd.concat thay thế
vertical_concat = pd.concat([series_x_21, series_y_21], axis=0)
print("\nThêm series y vào series x theo chiều dọc (Vertical concat):\n", vertical_concat)

# Thêm series y vào series x theo chiều ngang (horizontal)
horizontal_concat = pd.concat([series_x_21, series_y_21], axis=1)
print("\nThêm series y vào series x theo chiều ngang (Horizontal concat):\n", horizontal_concat)
```

#### Ví dụ Bài 15: Phép tính trên Series với dữ liệu nhập từ bàn phím

```python
import pandas as pd

def get_series_input(name):
    input_str = input(f"Nhập các số cho {name}, cách nhau bởi dấu phẩy: ")
    data = [float(x.strip()) for x in input_str.split(',')]
    return pd.Series(data)

sr_a = get_series_input("Series A")
sr_b = get_series_input("Series B")

if len(sr_a) == len(sr_b):
    results = pd.DataFrame({
        'A': sr_a,
        'B': sr_b,
        'A + B': sr_a + sr_b,
        'A - B': sr_a - sr_b,
        'A * B': sr_a * sr_b,
        'A / B': sr_a / sr_b
    })
    display(results)
else:
    print("Lỗi: Hai Series phải có cùng độ dài để thực hiện phép tính.")
```
